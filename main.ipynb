```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
events_df = pd.read_csv("data/events.csv")
```

```python
events_df.head()
```

```python
events_df.columns
```

```python
fig = plt.figure(figsize=(12, 10))
sns.lineplot(data=events_df, x="id", y="highest_score")
plt.xticks(events_df["id"])
plt.xlabel("Game Week")
plt.ylabel("Highest Score")
```

```python
fig = plt.figure(figsize=(12, 10))
sns.lineplot(data=events_df, x="id", y="transfers_made")
plt.xticks(events_df["id"])
plt.xlabel("Game Week")
plt.ylabel("Transfers Made")
```
